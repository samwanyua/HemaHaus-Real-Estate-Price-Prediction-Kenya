```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/processed/clean_data.csv", index_col=0)
data.head()
```

```python
# Drop unneeded columns for EDA and modeling
columns_to_drop = ['price_numeric', 'page', 'url', 'property_type','inserted_at', 'has_size', 'source', 'address']
data = data.drop(columns=columns_to_drop)

# Display remaining columns
data.columns
```

```python
# Round and cap at 5
data['bedrooms'] = np.round(data['bedrooms']).clip(upper=5).astype(int)
data['bathrooms'] = np.round(data['bathrooms']).clip(upper=5).astype(int)

# Check the cleaned unique values
print("Bedrooms:", sorted(data['bedrooms'].unique()))
print("Bathrooms:", sorted(data['bathrooms'].unique()))
```

```python
# Check basic price stats
print(data['price'].describe())

# Check zero or negative prices
print("Zero or negative prices:")
print(data[data['price'] <= 0])

# Check top 10 highest priced listings
print("Top 10 most expensive listings:")
print(data.sort_values('price', ascending=False).head(10))

# Check bottom 10 least expensive listings
print("Bottom 10 cheapest listings:")
print(data.sort_values('price').head(10))
```

```python
# Set max realistic price threshold
MAX_REALISTIC_PRICE = 200_000_000

# Filter the dataset
data = data[data['price'] <= MAX_REALISTIC_PRICE]
```

### Data visualization
1. Distribution of property prices

```python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=100, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price (KES)')
plt.ylabel('Frequency')
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.tight_layout()
plt.show()
```

 2. Price vs. Estimated Size (sqm)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='estimated_size_sqm', y='price', data=data, alpha=0.5)
plt.title('Price vs. Estimated Property Size')
plt.xlabel('Estimated Size (sqm)')
plt.ylabel('Price (KES)')
plt.xlim(0, 1000)
plt.ylim(0, 200_000_000)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y):,}'))
plt.tight_layout()
plt.show()
```

3. Average Price by Number of Bedrooms

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='bedrooms', y='price', data=data)
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Average Price (KES)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y):,}'))
plt.tight_layout()
plt.show()
```

4. Top 10 Most Frequent Locations by Price

```python
top_locations = data['location'].value_counts().nlargest(10).index
plt.figure(figsize=(10, 6))
sns.boxplot(x='location', y='price', data=data[data['location'].isin(top_locations)])
plt.title('Price Distribution in Top 10 Most Frequent Locations')
plt.xlabel('Location')
plt.ylabel('Price (KES)')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y):,}'))
plt.tight_layout()
plt.show()
```

5. Bathrooms vs. Price

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='bathrooms', y='price', data=data)
plt.title('Price Distribution by Number of Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price (KES)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y):,}'))
plt.tight_layout()
plt.show()
```

6. Top 10 Locations by Listing Count

```python
plt.figure(figsize=(12, 6))
top_locations = data['location'].value_counts().nlargest(10)
sns.barplot(x=top_locations.index, y=top_locations.values)
plt.title('Top 10 Locations by Number of Listings')
plt.ylabel('Number of Properties')
plt.xlabel('Location')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

7. Price Distribution by Location (Box Plot)

```python
plt.figure(figsize=(14, 6))
top_locs = data['location'].value_counts().nlargest(6).index
subset = data[data['location'].isin(top_locs)]
sns.boxplot(x='location', y='price', data=subset)
plt.title('Price Distribution by Location')
plt.ylabel('Price (KES)')
plt.xlabel('Location')
plt.yscale('log')  # In case of wide price range
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y):,}'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

8. Listings by Bedroom Count (Bar Plot)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='bedrooms', data=data)
plt.title('Distribution of Listings by Bedroom Count')
plt.xlabel('Bedrooms')
plt.ylabel('Number of Listings')
plt.tight_layout()
plt.show()
```

9. Listings by Bathroom Count

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='bathrooms', data=data)
plt.title('Distribution of Listings by Bathroom Count')
plt.xlabel('Bathrooms')
plt.ylabel('Number of Listings')
plt.tight_layout()
plt.show()
```

10. Price Range Bucket Chart

```python
bins = [0, 5e6, 10e6, 20e6, 50e6, 100e6, 200e6]
labels = ['<5M', '5–10M', '10–20M', '20–50M', '50–100M', '>100M']
data['price_range'] = pd.cut(data['price'], bins=bins, labels=labels)

plt.figure(figsize=(10, 5))
sns.countplot(x='price_range', data=data)
plt.title('Properties by Price Range')
plt.xlabel('Price Range (KES)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

11. Bedrooms vs Bathrooms Heatmap

```python
heat = data.pivot_table(index='bedrooms', columns='bathrooms', values='price', aggfunc='count', fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(heat, annot=True, fmt='d', cmap='BuPu')
plt.title('Number of Listings by Bedroom and Bathroom Count')
plt.xlabel('Bathrooms')
plt.ylabel('Bedrooms')
plt.tight_layout()
plt.show()
```

12. Price vs Bedrooms Scatter Plot

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='bedrooms', y='price', data=data)
plt.title('Price vs Bedroom Count')
plt.xlabel('Bedrooms')
plt.ylabel('Price (KES)')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y):,}'))
plt.tight_layout()
plt.show()
```
