```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/processed/clean_data.csv", index_col=0)
data.head()
```

```python
# Drop unneeded columns for EDA and modeling
columns_to_drop = ['price_numeric', 'page', 'url', 'property_type','inserted_at', 'has_size', 'source', 'address']
data = data.drop(columns=columns_to_drop)

# Display remaining columns
data.columns
```

```python
# Round and cap at 5
data['bedrooms'] = np.round(data['bedrooms']).clip(upper=5).astype(int)
data['bathrooms'] = np.round(data['bathrooms']).clip(upper=5).astype(int)

# Check the cleaned unique values
print("Bedrooms:", sorted(data['bedrooms'].unique()))
print("Bathrooms:", sorted(data['bathrooms'].unique()))
```

### Data visualization
1. Price distribution

```python
threshold = data['price'].quantile(0.99)
filtered_prices = data[data['price'] <= threshold]['price']

plt.figure(figsize=(10, 5))
sns.histplot(filtered_prices, bins=50, kde=True)
plt.title("Distribution of Property Prices (Excl. Top 1%)")
plt.xlabel("Price (KES)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
